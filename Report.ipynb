# Dataset for Analysis

The dataset is from FiveThirtyEight called `nba-elo`. `nba-elo` is discussed in a FiveThirtyEight article called [The Complete History Of The NBA](https://projects.fivethirtyeight.com/complete-history-of-the-nba/#warriors) which accounts for every NBA game played since its inception in 1947. This `nba-elo` dataset is open and can be found on [GitHub](https://github.com/fivethirtyeight/data/tree/master/nba-elo). Specific game information in the dataset on GitHub is from [Basketball-Reference.com](https://www.basketball-reference.com/). When it comes to licensing and terms of use for these datasets, they are "available under the [Creative Commons Attribution 4.0 International license](https://creativecommons.org/licenses/by/4.0/), and the code is available under the [MIT license](https://opensource.org/license/mit/)".
Each row in the dataset represent one game between two teams. The row also the date, team names, before and after elo's for each team, and points total for each team. The dataset is appropriate for addressing my research because they present game data that I can grab and compare with each team's elo. 
In regards to possible ethicial considerations in this dataset, I did not find any piece of information that would impede on personal information and privacy. I assume the current elo for every team to be valid and final based on the assumption that all game scores are final.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('nbaallelo.csv')
```

```python
no_dupes = df[df['_iscopy'] == 0]
```

```python
celtics = no_dupes[no_dupes['team_id'].str.contains('BOS')]
```

```python
lakers = no_dupes[no_dupes['team_id'].str.contains('LAL')]
```

# Results

From my data analysis of both teams, I have looked into how much of an impact *Bill Russell* and *Kobe Bryant* made in terms of their respective teams they played in and in their NBA seasons where they won championships. As they are both ranked first according to Bleacher Report<sup>[2], [3]</sup>, I began with the presumption that these players have made great impact on their team's elo.

## Bill Russell
Over the course of his 13-season stint with the Celtics, Bill Russell won **11** championships. In those 11 championship seasons, the Celtics averaged a ***7.932*** elo rating change with Bill Russell in the team along with a ***178.848*** net change in elo.

## Kobe Bryant
Kobe Bryant won **5** championships with the Lakers and saw relatively similar net change in the Laker's elo when he was with them in their championship seasons, which was ***143.528***. When the Lakers had Kobe Bryant in those championship seasons, there was a average of ***28.706*** elo rating change.

## Further Findings

When examining the Celtics and Lakers' final championship season from both Bill Russell and Kobe Bryant, both showed a significant drop in their elo over helfway through the season and jumps back up leading to the playoffs and the finals.

```python
celtics_1969 = celtics[celtics['year_id'] == 1969][['elo_n', 'date_game']]
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(celtics_1969['date_game'], celtics_1969['elo_n']);
c_69_labels = ['10/23/1968', '5/3/1969']
plt.xticks(c_69_labels)
plt.title("Celtics '69 Season")
plt.xlabel("Game Date")
plt.ylabel("Elo")
plt.show()
```

```python
lakers_2010 = lakers[lakers['year_id'] == 2010][['elo_n', 'date_game']]
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(lakers_2010['date_game'], lakers_2010['elo_n']);
l_10_labels = ['10/27/2009', '6/17/2010']
plt.xticks(l_10_labels)
plt.title("Lakers '10 Season")
plt.xlabel("Game Date")
plt.ylabel("Elo")
plt.show()
```
